<b> Methodology</b>

Teaching, research, fostering employability, and internationalization are the four pillars of a university's mission, and QS created its rankings to evaluate achievement in these areas. The methodology it employs evaluates universities using six metrics:
 - Academic Reputation (40%)
 - Employer Reputation (10%)
 - Faculty/Student Ratio (20%)
 - Citations per faculty (20%)
 - International Faculty Ratio (5%)
 - International Student Ratio (5%)

More information about the methodology can be found [here](https://www.topuniversities.com/qs-world-university-rankings/methodology "Please Visit")

<h2 style='text-align: center;'> Data Cleaning and Preprocessing</h2>

It is evident that there are several columns that are with null values, hence missing values. Columns like rank_display, score, city, type,  research_output, student_faculty_ratio, international_students, size, faculty_count. Hence lets, look at the main figures or numbers that are missing from the columns.

```python
pd.DataFrame(Unidata.isnull().sum(), columns=['No. of Missing values'])
```

Need to check the percentage of the missing values, so we know how to treat them appropriately.

```python
missing_percentage = round(Unidata.isna().mean() * 100, 1)
pd.DataFrame(missing_percentage[missing_percentage > 0], columns=['% of Missing Values'])
```

```python
Unidata['score']
```

To deal with this missing values adequetely, we need to understand the correlation if there's any, between the missing numbers. The missingno package is what I used. Thia will be achieved by visually representing the correlation between the incomplete values.

```python
msno.heatmap(Unidata, figsize=(10, 6), fontsize=12);
```

Only the columns with missing values are included in the correlation heatmap. The more the missing values in one column depend on the missing values in the other, the higher the correlation. As we can see, the variables "faculty count," "student faculty ratio," and "international students" are significantly correlated. Other columns show little to no correlation, if any at all.
We can't operate with a university that lacks many of its qualities, so let's eliminate rows with more than four missing values since several columns have missing values.

```python
print(len(Unidata[Unidata.isnull().sum(axis=1) > 4]))
drop_index = Unidata[Unidata.isnull().sum(axis=1) > 4].index.tolist()

Unidata.drop(drop_index, inplace=True)
print('Rows with more than 4 null values have been dropped!')
```

We'll remove the "link" and "logo" columns since they include hyperlinks. Although the "score" column might be highly helpful for analysis, roughly 56% of its data are missing. I discovered that although more than 1,000 universities were ranked, the QS website only provided a score for the top 500 universities. Therefore, I'm also ignoring by removing this column.

```python
Unidata.drop(['link', 'logo', 'score'], axis=1, inplace=True)
```

```python
Unidata['research_output'] = Unidata['research_output'].replace('Very high', 'Very High')
Unidata['international_students'] = Unidata['international_students'].apply(lambda x: float(str(x).replace(',','')))
Unidata['faculty_count'] = Unidata['faculty_count'].apply(lambda x: float(str(x).replace(',','')))
Unidata['rank_display'] = Unidata['rank_display'].apply(lambda x: float(re.sub(r'\W+', '', str(x))))
```

```python
yearly_Ranking = Unidata['year'].value_counts().sort_values()
fig, ax = plt.subplots(figsize=(10,4), dpi=90)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.tick_params(bottom=False)
ax.get_yaxis().set_visible(False)

sns.countplot(data=Unidata, x='year');

# add values on top of each bar
ax.bar_label(ax.containers[0])

ax.set_xlabel('Year', fontsize=13);
fig.suptitle('Number of universities ranked over the years', fontsize=15)
```
